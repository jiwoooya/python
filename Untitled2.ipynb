```python
import pandas as pd
import matplotlib.pyplot as plt

# 가상의 데이터로 예시를 보여줍니다.
data = {'dates': ['2019-01', '2019-02', '2019-03', '2019-04', '2019-05', '2019-06'],
        'elecprodqty': [100, 120, 90, 110, 130, 100]}

# 데이터를 데이터프레임으로 변환합니다.
df = pd.DataFrame(data)

# 'dates' 열을 날짜 형식으로 변환합니다.
df['dates'] = pd.to_datetime(df['dates'], format='%Y-%m')

# 그래프를 그립니다.
fig, ax = plt.subplots(figsize=(16, 8))  # 그래프의 넓이를 16로 조정합니다.
ax.barh(df['dates'], df['elecprodqty'], color='pink', height=20)  # 막대의 넓이를 0.8로 조정합니다.

ax.set_xlabel('전기 사용량', fontweight='bold', fontsize=12)
ax.set_ylabel('발생월일시', fontweight='bold', fontsize=12)
ax.set_title('2019년 전기 사용량', fontweight='bold', fontsize=12)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 가상의 데이터로 예시를 보여줍니다.
data = {'categories': ['2017', '2018', '2019', '2020'],
        'values': [30, 20, 15, 35]}

# 데이터를 데이터프레임으로 변환합니다.
df = pd.DataFrame(data)

# 원형 그래프를 그립니다.
plt.figure(figsize=(8, 8))
plt.pie(df['values'], labels=df['categories'], autopct='%1.1f%%', colors=['pink', 'lightblue', 'lightgreen', 'orange'])

plt.title('연도별 전기사용량 분포', fontweight='bold', fontsize=12)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 가상의 데이터로 예시를 보여줍니다.
data = {'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
        'electricity_usage': [100, 120, 90, 110, 130, 100]}

# 데이터를 데이터프레임으로 변환합니다.
df = pd.DataFrame(data)

# 롤리팝 차트를 그립니다.
plt.figure(figsize=(8, 6))
plt.stem(df['month'], df['electricity_usage'], markerfmt='o', linefmt='g-', basefmt=' ')

plt.xlabel('월', fontweight='bold', fontsize=12)
plt.ylabel('전기 사용량', fontweight='bold', fontsize=12)
plt.title('월별 전기 사용량', fontweight='bold', fontsize=12)

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 가상의 데이터로 예시를 보여줍니다.
data = {'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
        'electricity_usage': [100, 120, 90, 110, 130, 100],
        'population': [30, 50, 20, 40, 60, 30]}

# 데이터를 데이터프레임으로 변환합니다.
df = pd.DataFrame(data)

# 버블 차트를 그립니다.
plt.figure(figsize=(8, 6))
plt.scatter(df['month'], df['electricity_usage'], s=df['population']*10, c='blue', alpha=0.5)

plt.xlabel('월', fontweight='bold', fontsize=12)
plt.ylabel('전기 사용량', fontweight='bold', fontsize=12)
plt.title('월별 전기 사용량', fontweight='bold', fontsize=12)

plt.show()
```
